```python
class environmnent:
  def __init__(self,orginal,demo):
    self.steps = 2
    self.orginal = orginal
    self.error_1 = np.inf * np.ones(orginal.shape)
    self.error_2 = np.zeros(orginal.shape)
    self.demo = demo
  def error(self):
    self.error_2 = (self.orginal - self.demo)
    return self.error_2
  
  def get_observation(self):
    self.error_2 = self.error()
    return self.demo
    

  def get_action(self):
    #INCREMENT or decrement ALL cell BY +1 or -1

    demo_x = self.demo + np.ones(self.demo.shape)
    self.demo = demo_x
    obs_x = self.get_observation()
    reward_x = self.action()

    demo_y = self.demo - np.ones(self.demo.shape)
    self.demo = demo_y
    obs_y = self.get_observation()
    reward_y = self.action()
    
    if reward_x > reward_y:
      self.demo = demo_x
      return obs_x
    else:
      self.demo = demo_y
      return obs_y

  def is_done(self):
    return self.steps == 0

  def action(self):

    if (self.error_2 <= self.error_1).all():
      reward = 1
    elif(self.error_2[:,:1] <= self.error_1[:,:1]).all():
      reward = 0.6
    elif(self.error_2[:,1:2] <= self.error_1[:,1:2]).all():
      reward = 0.5
    elif(self.error_2[:,2:3] <= self.error_1[:,2:3]).all():
      reward = 0.3
    elif(self.error_2[:,3:] <= self.error_1[:,3:]).all():
      reward = 0.2
    elif(self.error_2[:,:3] <= self.error_1[:,:3]).all():
      reward = 0.6
    if (self.error_2 >= self.error_1).all():
      reward = 0
    
    self.error_1 = self.error_2
    self.steps -= 1

    return reward

```

```python
class Agent:
  def __init__(self):
     self.total_reward = 0
  
  def steps (self, env: environmnent):
    current_obs = env.get_observation()
    #print(current_obs)
    actions = env.get_action()
    #print(len(actions))
    reward = env.action()
    print(reward)
    self.total_reward += reward
```
